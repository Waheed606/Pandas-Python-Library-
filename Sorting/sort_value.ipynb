## sorting the value

```python
import pandas as pd

data =pd.read_csv("Overseas.csv")
```

# Working on series 

```python
series_sort=data['Continent'].value_counts()
```

```python
series_sort
```

## Sorting in Ascending Order

```python
series_sort.sort_values()
```

## Sorting in Descending Order

```python
series_sort.sort_values(ascending=False)
```
